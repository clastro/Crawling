```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
```

```python
path = "C:/chromedriver_win32/chromedriver.exe" #크롬 드라이버 파일 다운로드 후 경로 설정해야 함
driver = webdriver.Chrome(path)
```

```python
#Iframe 안에 있는 유효한 url 주소를 찾아내서 입력해야 한다
#여기까지 작업 필요
url = 'https://chp-dashboard.geodata.gov.hk/covid-19/apps/Tab1_Mobile_Eng/'
driver.get(url)
```

```python
# 필요하면 soup사용
#soup = BeautifulSoup(html.text, 'html.parser')
```

```python
text = driver.find_elements_by_class_name("list-item.dgrid-row") #원하는 정보가 담긴 리스트
#text = driver.find_elements_by_class_name("dgrid-content")
```

```python
case_list = []
```

```python
#text에 한 개에 80~160개 정도 리스트가 들어감
#두 번 스크롤 변화 시키면서 데이터 저장하는 추가 코드 필요

for n, answer in enumerate(text):
        
        texts = answer.find_elements_by_tag_name('p')
            
        t = ""     
        
        for i in texts:             
            
            t += i.text
        case_list.append(t)
        print(t)
        
    
```

```python
case_list
```

```python
data = pd.DataFrame(case_list)
```

```python
final_data = data.drop_duplicates(keep='first') #중복제거
```

```python
final_data.to_csv("list_crawl.csv",encoding="utf-8-sig")
```

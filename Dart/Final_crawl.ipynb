```python
k = 0
for url in df_rcp_new["report_6"]:
    string = ''
    if(url == "modified"):
        df_rcp_new.loc[k,"bosang"] = 0
        k+=1
    else:
        html = urlopen(url)
        time.sleep(1)
        bsObject = BeautifulSoup(html, "html.parser")
        time.sleep(1)
        text = bsObject.find_all('p')
        time.sleep(1)
        for x in text:
            string += str(x.get_text())
        switch = string.find("보상위원회")
        time.sleep(1)
        if(switch>0):
            df_rcp_new.loc[k,"bosang"] = 1
        k+=1    
    if(k%20 ==0):
        df_rcp_new.to_csv("./result/bosang/7th_"+str(k)+"save.csv",encoding="utf-8-sig")
        time.sleep(1)
    
```
